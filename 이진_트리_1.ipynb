```python
#전위순회
def preorder(node):
  if node == None:
    return
  print(node.data, end = ' ' )
  preorder(node.left)
  preorder(node.right)

#중위순회
def inorder(node):
  if node == None:
    return
  inorder(node.left)
  print(node.data, end = ' ' )
  inorder(node.right)

#후위순회
def postorder(node):
  if node == None:
    return
  inorder(node.left)
  inorder(node.right)
  print(node.data, end = ' ' )

print("전위순위", end = "")
preorder(node1)
print("끝")


print("중위순위", end = "")
inorder(node1)
print("끝")

print("후위순위", end = "")
postorder(node1)
print("끝")
```

# **이진 탐색 트리의 일반 구현**

```python
memory = []
root = None
```

```python
nameAry = ["블랙핑크","레드벨렛","마마무","걸스데이","트와이스"]
```

```python
#root 생성
node = TreeNode()
node.data = nameAry[0]
root = node
memory.append(node)
```

```python
for name in nameAry[1:]:

  node = TreeNode()
  node.data = name

  current = root
  while True:
    if(current.data < node.data):
      if current.left == None:
        current.left = node
        break
      current = current.left

    if(current.data > node.data):
      if current.right == None:
        current.right = node
        break
    current = current.right

    memory.append(node)

print("종료")
```

이진 탐색 트리의 검색 자동

```python
findName = "마마무"

current = root

while True:

  if current.data == findName:
    print(findName, "을 찾음")
    break

  elif current.data > findName:
    if current.right == None:
      print(findName, "을 찾을 수 없음")
      break
    current = current.right

  else:
    if current.left == None:
      print(findName, "을 찾을 수 없음")
      break
    current = current.left
```

이진 탐색 트리에서 데이터 삭제

- 리브 노드를 삭제하는 경우
- 자식 노드가 한개인 노드를 삭제하는 경우
- 자식 노드가 2개 이상인 노드를 삭제하는 경우

```python
deleteName = '마마무'

current = root
parent = None

while True:
  if deleteName == current.data:

    if current.left == None and current.right == None:
      if parent.left == current:
        parent.left = None
      else:
        parent.right = None
      del(current)

    elif current.left != None and current.right == None:
      if parent.left == current:
        parent.left = current.left
      else:
        parent.right = current.left
      del(current)

    elif current.left == None and current.right != None:
      if parent.left == current:
        parent.left = current.right
      else:
        parent.right = current.right
      del(current)

    print(current,"이(가) 삭제됨")
    break

  elif deleteName < current.data:
    if current.left == None:
      print(deleteName,"이(가) 트리에 없음(좌)")
      break
    parent = current
    current = current.left

  else:
    if current.right == None:
      print(deleteName,"이(가) 트리에 없음(우)")
      break
    parent = current
    current = current.right


```
