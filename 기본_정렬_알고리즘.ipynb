# **선택정렬**
: 여러 데이터 중에서 가장 작은 값을 뽑는 작동을 반복하여 값을 정렬하는 방식

- 앱력 개수가 커질수록 기하급수적으로 비교횟수가 증가하기 때문에 성능이 나쁜 알고리즘이다.

```python
def findMinIdx(ary):
  minIdx = 0
  for i in range(1, len(ary)):
    if (ary[minIdx] > ary[i]):
      minIdx = i
  return minIdx

before = [44,67,54,76,8,3]
after = []

for _ in range(len(before)):
  minPos = findMinIdx(before)
  after.append(before[minPos])
  del(before[minPos])

print(after)
```

```python
def selectionSort(ary):
  n = len(ary)
  for i in range(0,n-1):
    minIdx = i
    for k in range(i+1, n):
      if (ary[minIdx] > ary[k]):
        minIdx = k
    tmp = ary[i]
    ary[i] = ary[minIdx]
    ary[minIdx]=tmp

  return ary

dataAry = [1256,243,675,324,3245]
print("orgin ==> ", dataAry)
dataAry = selectionSort(dataAry)
print("after ==> ", dataAry)
```

```python
def findInsertIdx(ary, data):
  findIdx = -1
  for i in range(0,len(ary)):
    if (ary[i]> data):
      findIdx = i
      break

  if findIdx == -1:
    return len(ary)
  else:
    return findIdx

before = [123,645,7869,34,2,57,8]
after = []

print("정렬 전 ==>", before)

for i in range(len(before)):
  data = before[i]
  insPos = findInsertIdx(after, data)
  after.insert(insPos, data)

print("정렬 후 ==>", after)
```

```python
def insertionSort(ary):
  n = len(ary)
  for end in range(1,n):
    for cur in range(end, 0, -1):
      if(ary[cur-1] > ary[cur]):
        ary[cur -1], ary[cur] = ary[cur], ary[cur-1]
  return ary


## 전역 변수 선언 부분 ##

dataAry = [123,435,654,23,65,34,654]

## 메인 코드 부분 ##

print("정렬 전 ==>", dataAry)
dataAry = insertionSort(dataAry)
print("정렬 후 ==>", dataAry)
```

# **버블 정렬**



```python
def BubbleSort(ary):
  n = len(ary)
  for end in range(n-1,0,-1):
    for cur in range(0,end):
      if (ary[cur] > ary[cur+1]):
        ary[cur], ary[cur+1] = ary[cur+1], ary[cur]

    return ary

## 전역 변수 선언 부분

dataAry = [188,162,168,120,50,150,177,105]

## 메인 코드 부분

print("정렬 전 -->", dataAry)
dataAry = BubbleSort(dataAry)
print("정렬 전 -->", dataAry)
```

# **퀵 정렬**

기준을 하나 뽑은 후 기준보다 작은 그룹과 큰 그룹을 나누어 다시 각 그룹으로 정렬하는 방식.
이때 나눈 그룹을 다시 정렬하고자 재귀 호출을 하고, 각 그룹의 정렬이 완료되면 합치는 방식

```python
def quickSort(ary):

  n = len(ary)
  if n <= 1:
    return ary

  pivot = ary[n//2]
  leftAry, rightAry = [],[]

  for num in ary:
    if num < pivot:
      leftAry.append(num)
    elif num > pivot:
      rightAry.append(num)

  return quickSort(leftAry) + [pivot] + quickSort(rightAry)

## 전역 변수 선언 부분

dataAry = [188,162,168,120,50,150,177,105]

## 메인 코드 부분

print("정렬 전 -->", dataAry)
dataAry = quickSort(dataAry)
print("정렬 후 -->", dataAry)
```

```python
def quickSort(ary):

  n = len(ary)
  if n <= 1:
    return ary

  pivot = ary[n//2]
  leftAry, midAry, rightAry = [],[],[]

  for num in ary:
    if num < pivot:
      leftAry.append(num)
    elif num > pivot:
      rightAry.append(num)
    else: midAry.append(num)

  return quickSort(leftAry) + midAry + quickSort(rightAry)

## 전역 변수 선언 부분

dataAry = [188,162,168,120,50,150,90,168,177,105]

## 메인 코드 부분

print("정렬 전 -->", dataAry)
dataAry = quickSort(dataAry)
print("정렬 후 -->", dataAry)
```
